## Arithmetic Expressions Examples

Arithmetic expressions are fundamental in programming and data analysis. They involve the basic operations of addition, subtraction, multiplication, division, and more. Here are some examples of arithmetic expressions:

### Basic Operations

1. **Addition (`+`)**:
   - Example: \(3 + 2 = 5\)
   - Python Code: `3 + 2`

2. **Subtraction (`-`)**:
   - Example: \(5 - 2 = 3\)
   - Python Code: `5 - 2`

3. **Multiplication (`*`)**:
   - Example: \(4 \times 3 = 12\)
   - Python Code: `4 * 3`

4. **Division (`/`)**:
   - Example: \(10 / 2 = 5\)
   - Python Code: `10 / 2`

### Advanced Operations

1. **Exponentiation (`**`)**:
   - Example: \(2^3 = 8\)
   - Python Code: `2 ** 3`

2. **Floor Division (`//`)**:
   - Example: \(7 // 2 = 3\) (Division that rounds down to the nearest integer)
   - Python Code: `7 // 2`

3. **Modulus (`%`)**:
   - Example: \(7 \% 2 = 1\) (Remainder of the division)
   - Python Code: `7 % 2`

### Combining Operations

Arithmetic expressions can combine multiple operations using parentheses to dictate the order of operations (PEMDAS/BODMAS rules).

- Example: \((3 + 2) \times 4 = 20\)
- Python Code: `(3 + 2) * 4`

- Example: \(10 - 4 / 2 + 3 = 11\)
- Python Code: `10 - 4 / 2 + 3`
